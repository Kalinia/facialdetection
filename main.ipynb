```python
from matplotlib import pyplot as plt
from matplotlib.patches import Rectangle
from mtcnn.mtcnn import MTCNN
 
# draw an image with detected objects
def draw_image_with_boxes(filename, result_list):
    # load the image
    plt.xticks([])
    plt.yticks([])
    data = plt.imread(filename)
    # plot the image
    plt.imshow(data)
    # get the context for drawing boxes
    ax = plt.gca()
    # plot each box
    for result in result_list:
        # get coordinates
        x, y, width, height = result['box']
        # create the shape
        rect = Rectangle((x, y), width, height, fill=False, color='red')
        # draw the box
        ax.add_patch(rect)
    # show the plot
    plt.show()
    
    
def draw_faces_all(filename, result_list):
    # load the image
    data = plt.imread(filename)
    # plot each face as a subplot
    for i in range(len(result_list)):
        # get coordinates
        x1, y1, width, height = result_list[i]['box']
        x2, y2 = x1 + width, y1 + height
        # define subplot
        plt.subplot(1, len(result_list), i+1)
        plt.axis('off')
        # plot face
        
        plt.imshow(data[y1:y2, x1:x2])
    # show the plot
    plt.show()
    
    
def draw_faces_separate(filename, result_list):
    # load the image
    data = plt.imread(filename)
    # plot each face as a subplot
    for i in range(len(result_list)):
        # get coordinates
        plt.figure(i)
        x1, y1, width, height = result_list[i]['box']
        x2, y2 = x1 + width, y1 + height
        # define subplot
#         pyplot.subplot(1, len(result_list), i+1)
        plt.axis('off')
        # plot face
        
        plt.imshow(data[y1:y2, x1:x2])
        plt.show()
    # show the plot
    
```

```python
name = 'images/test13.jpg'

image = plt.imread(name)
nn = MTCNN()
faces = nn.detect_faces(image)

draw_image_with_boxes(name, faces)
```

```python
name = 'images/test3.jpg'

image = plt.imread(name)
nn = MTCNN()
faces = nn.detect_faces(image)

draw_image_with_boxes(name, faces)
draw_faces_all(name, faces)
draw_faces_separate(name, faces)
```
